```python
import pandas as pd
import numpy as np
```

```python
bee_data = pd.read_csv("../bee_data/machine_learning_dataset.csv")
bee_data
```

```python
bee_data.drop("Unnamed: 0", inplace=True, axis=1)
bee_data
```

### Train and Test Data

```python
from sklearn.model_selection import train_test_split
```

```python
X = bee_data.drop(["year", "lbs_of_honey", "state"], axis=1)
```

```python
y = bee_data["lbs_of_honey"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Standardisations

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
train_scaled = scaler.fit_transform(X_train)
```

```python
test_scaled = scaler.transform(X_test)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(train_scaled, y_train)
```

### Model Evaluation

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train, model.predict(train_scaled))
```

```python
accuracy_score(y_train, model.predict(test_scaled))
```

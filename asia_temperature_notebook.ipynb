```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import csv
import json
```

```python
temp_data = pd.read_csv("bee_data/city_temperature.csv", low_memory =False)
temp_data
```

```python
temp_data_filt = (temp_data['Country'] == 'US')
```

```python
all_us_data = temp_data[temp_data_filt]
all_us_data.head()
```

# Too Cold

```python
cold_days = all_us_data["AvgTemperature"] <= 41
# hot_days = all_us_data["AvgTemperature"] >100

cold_df = all_us_data[cold_days]
cold_df 
```

```python
indexed_errors = cold_df[cold_df['AvgTemperature'] == -99.0].index
# indexNames
# # Delete these row indexes from dataFrame
cold_df.drop(indexed_errors, inplace=True)
cold_df
```

```python
state_cold_days = cold_df.groupby(["Year", "State", "City"], as_index=False)[['AvgTemperature']]
state_cold_days.count()
```

# Too Hot

```python
hot_days = all_us_data["AvgTemperature"] > 95
# hot_days = all_us_data["AvgTemperature"] >100

hot_df = all_us_data[hot_days]
hot_df
```

```python
state_hot_days = hot_df.groupby(["Year", "State", "City"], as_index=False)[['AvgTemperature']]
state_hot_days.count()
```
